```python
import pandas as pd
import os

# --- Configuration ---
# Le nom de votre fichier Excel
EXCEL_FILE_NAME = "Book1.xlsx"
# Le chemin complet vers le fichier Excel
# Assurez-vous que ce fichier est dans le même dossier que votre script Python,
# sinon, ajustez le chemin.
EXCEL_FILE_PATH = os.path.join(os.path.dirname(__file__), EXCEL_FILE_NAME)

# Les noms exacts de vos feuilles Excel
SHEET_STOCK_ACTUEL = "StockActuel"
SHEET_VENTES = "Ventes"
SHEET_HISTORIQUE_APPRO = "HistoriqueApprovisionnement"

# --- Fonction pour lire les données ---
def load_data(file_path):
    """
    Charge les données depuis les différentes feuilles du fichier Excel.

    Args:
        file_path (str): Le chemin complet vers le fichier Excel.

    Returns:
        tuple: Un tuple contenant trois DataFrames pandas (stock_df, ventes_df, historique_appro_df).
               Retourne (None, None, None) si une erreur survient (fichier non trouvé, feuille manquante).
    """
    stock_df = None
    ventes_df = None
    historique_appro_df = None

    try:
        # On utilise pd.ExcelFile pour ouvrir le fichier une seule fois
        excel_file = pd.ExcelFile(file_path)

        # Lire chaque feuille dans un DataFrame
        if SHEET_STOCK_ACTUEL in excel_file.sheet_names:
            stock_df = excel_file.parse(SHEET_STOCK_ACTUEL)
            print(f"Feuille '{SHEET_STOCK_ACTUEL}' chargée avec succès. {len(stock_df)} lignes trouvées.")
        else:
            print(f"Erreur : Feuille '{SHEET_STOCK_ACTUEL}' non trouvée dans '{EXCEL_FILE_NAME}'.")
            return None, None, None # Retourne None pour indiquer une erreur

        if SHEET_VENTES in excel_file.sheet_names:
            ventes_df = excel_file.parse(SHEET_VENTES)
            print(f"Feuille '{SHEET_VENTES}' chargée avec succès. {len(ventes_df)} lignes trouvées.")
        else:
            print(f"Erreur : Feuille '{SHEET_VENTES}' non trouvée dans '{EXCEL_FILE_NAME}'.")
            return None, None, None # Retourne None pour indiquer une erreur

        if SHEET_HISTORIQUE_APPRO in excel_file.sheet_names:
            historique_appro_df = excel_file.parse(SHEET_HISTORIQUE_APPRO)
            print(f"Feuille '{SHEET_HISTORIQUE_APPRO}' chargée avec succès. {len(historique_appro_df)} lignes trouvées.")
        else:
            print(f"Erreur : Feuille '{SHEET_HISTORIQUE_APPRO}' non trouvée dans '{EXCEL_FILE_NAME}'.")
            return None, None, None # Retourne None pour indiquer une erreur

        return stock_df, ventes_df, historique_appro_df

    except FileNotFoundError:
        print(f"Erreur : Le fichier '{EXCEL_FILE_NAME}' n'a pas été trouvé au chemin spécifié : {file_path}")
        return None, None, None # Retourne None pour indiquer l'erreur

    except Exception as e:
        print(f"Une erreur inattendue s'est produite lors de la lecture du fichier Excel : {e}")
        return None, None, None # Retourne None pour indiquer l'erreur

# --- Exemple d'utilisation ---
if __name__ == "__main__":
    # Cette partie s'exécute uniquement si vous lancez ce script directement
    print(f"Tentative de chargement des données depuis : {EXCEL_FILE_PATH}")
    stock_data, ventes_data, historique_appro_data = load_data(EXCEL_FILE_PATH)

    if stock_data is not None:
        print("\n--- Aperçu des données Stock Actuel ---")
        print(stock_data.head()) # Affiche les 5 premières lignes

        print("\n--- Aperçu des données Ventes ---")
        print(ventes_data.head())

        print("\n--- Aperçu des données Historique Approvisionnement ---")
        print(historique_appro_data.head())

    else:
        print("\nLe chargement des données a échoué. Veuillez vérifier le nom du fichier, son chemin et les noms des feuilles.")
```
